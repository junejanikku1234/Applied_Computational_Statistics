```python
import numpy as np
import matplotlib.pyplot as plt

def implement_clt(population, sample_sizes, num_samples=1000):
  """
  Demonstrates the Central Limit Theorem by sampling from a given population
  and plotting the distribution of sample means for different sample sizes.

  Args:
    population (list or numpy.ndarray): The population data to sample from.
    sample_sizes (list of int): A list of different sample sizes to test.
    num_samples (int): The number of sample means to calculate for each sample size.
  """

  population_mean = np.mean(population)
  print(f"Population Mean: {population_mean:.2f}")

  plt.figure(figsize=(12, 6))

  for i, n in enumerate(sample_sizes):
    sample_means = []
    for _ in range(num_samples):
      samples = np.random.choice(population, size=n, replace=False) # Sampling without replacement
      sample_mean = np.mean(samples)
      sample_means.append(sample_mean)

    plt.subplot(1, len(sample_sizes), i + 1)
    plt.hist(sample_means, bins=20, density=True, alpha=0.7, color='skyblue')
    plt.title(f'Sample Size (n) = {n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

    # Overlay the mean of the sample means
    mean_of_means = np.mean(sample_means)
    plt.axvline(mean_of_means, color='red', linestyle='dashed', linewidth=1, label=f'Mean of Means: {mean_of_means:.2f}')
    plt.legend()

  plt.suptitle('Demonstration of the Central Limit Theorem', y=1.02)
  plt.tight_layout()
  plt.show()

# Define the population (uniform distribution between 0 and 10)
population = np.linspace(0, 10, 1000) # Creating a dense representation of the uniform distribution

# Define the sample sizes to test
sample_sizes_to_test = [5, 30]

# Implement the Central Limit Theorem
implement_clt(population, sample_sizes_to_test)
```
