```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform

# Parameters for Normal Distribution
mu = 50
sigma = 10
x_norm = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_norm = norm.pdf(x_norm, mu, sigma)
cdf_norm = norm.cdf(x_norm, mu, sigma)

# Plot Normal Distribution
plt.figure(figsize=(10, 5))
plt.plot(x_norm, pdf_norm, label='Normal PDF', color='blue')
plt.plot(x_norm, cdf_norm, label='Normal CDF', color='green')
plt.title("Normal Distribution")
plt.xlabel("X")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()

# Parameters for Uniform Distribution
a, b = 2, 8
x_uni = np.linspace(a - 1, b + 1, 1000)
pdf_uni = uniform.pdf(x_uni, a, b - a)
cdf_uni = uniform.cdf(x_uni, a, b - a)

# Plot Uniform Distribution
plt.figure(figsize=(10, 5))
plt.plot(x_uni, pdf_uni, label='Uniform PDF', color='purple')
plt.plot(x_uni, cdf_uni, label='Uniform CDF', color='orange')
plt.title("Uniform Distribution")
plt.xlabel("X")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```
