```python
import matplotlib.pyplot as plt

# Define the PMF for a fair six-sided die
def pmf_fair_die(x):
    if x in [1, 2, 3, 4, 5, 6]:
        return 1/6
    else:
        return 0

# Possible outcomes of the die roll
outcomes = [1, 2, 3, 4, 5, 6]

# Calculate PMF values for each outcome
pmf_values = [pmf_fair_die(x) for x in outcomes]

# Plot the PMF
plt.stem(outcomes, pmf_values, basefmt=" ")
plt.title("Probability Mass Function (PMF) for a Fair Six-Sided Die")
plt.xlabel("Outcome (X)")
plt.ylabel("Probability P(X=x)")
plt.xticks(outcomes)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print PMF values
print("Outcome (X) | Probability P(X=x)")
print("---------------------------------")
for x, p in zip(outcomes, pmf_values):
    print(f"{x:11} | {p:.4f}")
```
