# Data Presentation of Qualitative Data

**Bar Chart**
A bar plot uses rectangular bars to represent data categories, with bar length or height proportional to their values. It compares discrete categories, with one axis for categories and the other for values.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({
    'Flavor': ['Chocolate', 'Vanilla', 'Mint', 'Swirl', 'Nut'],
    'Orders': [1500,  670,  950,  450,   75]
})
df.plot(kind='bar', x='Flavor', y='Orders',legend = None, title='Ice Cream Sales',xlabel='Flavor', ylabel='Orders')
plt.show()
```

Pie Chart

A pie chart is a circular depiction of data where the area of the whole pie represents 100% of
the data and slices of the pie represent a percentage breakdown of the sublevels.

```python
df.plot(kind='pie', labels=df['Flavor'], y='Orders', legend=None, autopct='%.2f%%', title='Ice Cream Sales')
plt.tight_layout()
plt.show()
```

Pareto Chart

 Pareto chart is a special kind of chart where the bars are ordered by frequency counts from highest to lowest. These charts are often used to identify areas to focus on first in process improvement.

```python
# Sort values by 'Orders' in descending order
df = df.sort_values(by='Orders', ascending=False)

# Extract x and y values
x = df['Flavor'].values  # Flavors
y = df['Orders'].values  # Orders

# Calculate weights (proportion of each order) and cumulative sum
weights = y / y.sum()        # Percentage contribution
cumsum = weights.cumsum()    # Cumulative percentage

# Plot bar chart for Orders
fig, ax1 = plt.subplots()
ax1.bar(x, y, color='skyblue')
ax1.set_xlabel('Flavor')  # x-axis label
ax1.set_ylabel('Orders', color='blue')  # y-axis label for the left axis

# Add a line plot for the cumulative percentage
ax2 = ax1.twinx()  # Create a secondary y-axis
ax2.plot(x, cumsum, '-ro', alpha=0.7)  # Plot cumulative percentage line
ax2.set_ylabel('Cumulative Percentage', color='red')
ax2.tick_params('y', colors='red')

# Add percentage labels to the cumulative percentage line
for i, txt in enumerate(cumsum):
    ax2.annotate(f'{txt:.0%}', (x[i], cumsum[i]), fontweight='bold', ha='center', va='bottom')

# Add a title
plt.title('Pareto Chart of Ice Cream Orders')

# Adjust layout and display the chart
plt.tight_layout()
plt.show()
```

Box-Plot

A Box plot is a way to visualize the distribution of the data by using a box and some vertical lines. It is known as the whisker plot. The data can be distributed between five key ranges, which are as follows:

    Minimum: Q1-1.5*IQR
    1st quartile (Q1): 25th percentile
    Median:50th percentile
    3rd quartile(Q3):75th percentile
    Maximum: Q3+1.5*IQR

Here IQR represents the InterQuartile Range which starts from the first quartile (Q1) and ends at the third quartile (Q3).

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame([[10, 20, 30, 40], [7, 14, 21, 28], [15, 15, 8, 12],
                   [15, 14, 1, 8], [7, 1, 1, 8], [5, 4, 9, 2]],
                  columns=['Apple', 'Orange', 'Banana', 'Pear'],
                  index=['Basket1', 'Basket2', 'Basket3', 'Basket4',
                         'Basket5', 'Basket6'])

box = plt.boxplot(df, patch_artist=True)

colors = ['blue', 'green', 'purple', 'tan', 'pink', 'red']

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

x = [[1.2, 2.3, 3.0, 4.5],
     [1.1, 2.2, 2.9, 5.0]]

df = pd.DataFrame(x, index=['Apple', 'Orange'])
df.T.boxplot()

plt.subplots_adjust(bottom=0.25)

plt.show()

```
